```python
## Converting everything to seconds
def converttoseconds(time_var):
     hh,mm,ss=time_var.split(':')
     return int(hh)*3600+int(mm)*60+int(ss)
```

```python
avg_response_timeinseconds=[]
#df.drop('Response_time_in_Seconds',axis=1,inplace=True
avg_response_timeinseconds
```

```python
for i in df['Average Response Time']:
    avg_response_timeinseconds.append(converttoseconds(i))
df['AvgResponse_time_in_Seconds']=avg_response_timeinseconds
```
