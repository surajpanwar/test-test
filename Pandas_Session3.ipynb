```python
data4={
    'emp_id': [101,102,103,104],
    'salary': [234,345,200,400],
    'pf': [23,34,56,78]
    
}
```

```python
data5={
    'emp_id': [101,102,104,105],
    'mob_no': [3434,5656,767,878],
    'house_nu': [345,345,345,56]
    
}
```

```python
df4=pd.DataFrame(data4)
df4
```

```python
df5=pd.DataFrame(data5)
df5
```

```python
pd.merge(df4,df5)
```

```python
## By default it will perform inner merge takes all the common values in merged table
```

```python
pd.merge(df4,df5,how='left')
## merge(leftdf, rightdf, how)
```

```python
pd.merge(df4,df5,how='right')
```

```python
## Missing values will be filled by NaN
```

```python
pd.merge(df4,df5,how='right',on="emp_id")
```

```python
df_sales=pd.read_csv("sales_data_final.csv")
```

```python
df_sales.head(2)
```

```python
def profit_flag(a):
    if a>=0:
        return 'Positive'
    else:
        return 'Negative'
```

```python
 profit_flag(67)
```

```python
 profit_flag(-4)
```

```python
df_sales['flag_profit']=df_sales['profit'].apply(profit_flag)
```

```python
df_sales.head()
```

```python
def lenght_name(name):
    return len(name)
```

```python
df_sales['Len_CustomerName']=df_sales['customer_name'].apply(len)
```

```python
df_sales.head(2)
```

```python
## Quantity Flag
```

```python
def quantity_flag(a):
    if a<10:
        return 'Low'
    elif a>10 and a<20:
        return 'Medium'
    else:
        return 'High'
```

```python
df_sales['flag_quantity']=df_sales['quantity'].apply(quantity_flag)
```

```python
df_sales.head()
```

```python
df_sales['square_quantity']=df_sales['quantity'].apply(lambda x: x*x)
```
